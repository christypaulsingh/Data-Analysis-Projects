# Correlation

```python
#Numerical columns
num_col=df.corr()
num_col
```

```python
sns.set(rc={'figure.figsize':(8,5)})
sns.heatmap(num_col,annot=True,)
plt.title("Numerical Columns Correlation")
```

```python
#Positive Correlation:
sns.lmplot(x='Sales',y='Profit',data=df)
plt.title("Positive Correlation - Sales and Profit")
```

*A positive correlation between sales and profit indicates that as sales increase, profits also rise. This suggests a direct and proportional relationship, where higher sales contribute to greater profitability for a business.*

```python
#Negative Correlation:
sns.lmplot(x='Profit',y='Discount',data=df)
plt.title("Negative Correlation - Profit and Discount")
```

*A negative correlation between discount and profit suggests that as discounts increase, profits tend to decrease. Focusing more on discounting may attract customers but could potentially erode profitability.*

```python
#Zero Correlation:
sns.lmplot(x='Quantity',y='Discount',data=df)
plt.title("Zero Correlation - Quantity and Discount")
```

*A zero correlation between quantity and discount indicates that changes in the quantity of a product purchased are not associated with any consistent changes in the discount applied to that product. In other words, the two variables do not show a linear relationship or influence each other.*

# Outliers

```python
#Find the outliers
sns.set(rc={'figure.figsize':(10,5)})
sns.boxplot(x='Sales',data=df)
plt.title("Boxplot-Sales Data")
```

```python
#Sales data distribution/ statistics:
df['Sales'].describe()
```

# IQR Method

```python
q1=df.Sales.quantile(.25)
q3=df.Sales.quantile(.55)
Iqr=q3-q1
lower_limit=q1-1.5*Iqr
upper_limit=q1+1.5*Iqr
```

# Finding Outliers

```python
df[df.Sales<upper_limit]
```

```python
df[df.Sales>lower_limit]
```

# Remove outliers

```python
new_df=df[(df.Sales>lower_limit)&(df.Sales<upper_limit)]
print("old data:",len(df))
print("new data:",len(new_df))
print("No.of.outliers:",len(df)-len(new_df))
```

```python
new_df.shape
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='Sales',data=df)
plt.xlabel("Old Sales")

plt.suptitle("Old Sales data and New Sales data")
plt.subplot(1,2,2)
sns.boxplot(x='Sales',data=new_df)
plt.xlabel("New Sales")
```

```python
new_df
```

*After removing all outliers from the dataset, the analysis becomes more robust and representative of the central trend, enhancing the reliability of statistical conclusions.*

# Univariate Analysis

```python
# Segment:
sns.set(rc={'figure.figsize':(5,5)})
seg=sns.countplot(x='Segment',data=new_df)
plt.title('Count of Segment')

for i in seg.containers:
    seg.bar_label(i)    
```

```python
#Category:
sns.set(rc={'figure.figsize':(5,5)})
category=sns.countplot(x='Category',data=new_df)
plt.title('Count of Category')


for i in category.containers:
    category.bar_label(i) 
```

```python
#Sub-Category:
sns.set(rc={'figure.figsize':(20,7)})
sub_category=sns.countplot(x='Sub-Category',data=new_df)
plt.title('Count of Sub-category')


for i in sub_category.containers:
    sub_category.bar_label(i) 
```

```python
#Region:
sns.set(rc={'figure.figsize':(10,5)})
x=new_df['Region'].value_counts()
plt.pie(x.values,labels=x.index,autopct='%1.2f%%',explode=[0.2,0,0,0])
plt.legend()
plt.show()
```

```python
#State:
sns.set(rc={'figure.figsize':(15,7)})
state=sns.countplot(x='State',data=new_df)
plt.xticks(rotation=90)
plt.title('Count of State')


for i in state.containers:
    state.bar_label(i) 
```

# Bivariate Analysis

```python
#1.Which region has a higher number of sales
sns.set(rc={'figure.figsize':(10,5)})
state=new_df.groupby('Region').sum().plot(kind='pie',y='Sales',autopct='%2.1f%%',title='Which region has a higher number of sales')
state
```

```python
#2.Which state has the highest number of sales on top state:
sns.set(rc={'figure.figsize':(12,5)})
state=new_df.groupby('State',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(10)
sns.barplot(x='State',y='Sales',data=state)
plt.title('Which state has the highest number of sales on top 10 State')
```

```python
#3.Which City has the highest number of sales on top state:
sns.set(rc={'figure.figsize':(15,5)})
city=new_df.groupby('City',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(10)
sns.barplot(x='City',y='Sales',data=city)
plt.title('Which city has the highest number of sales on top 10 city')
```

```python
#4.Which Category has the highest number of sales on top state:
sns.set(rc={'figure.figsize':(5,5)})
Category=new_df.groupby('Category',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
sns.barplot(x='Category',y='Sales',data=Category)
plt.title('Which Category has the highest number of sales')
```

```python
#5.Which Sub-Category has the highest number of sales on top 10 products:
sns.set(rc={'figure.figsize':(15,5)})
SCategory=new_df.groupby('Sub-Category',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(10)
sns.barplot(x='Sub-Category',y='Sales',data=SCategory)
plt.title('Which Sub-Category has the highest number of sales on top 10 products')
```

```python
#6.Discount and category:
sort=new_df.sort_values('Discount',ascending=False)
sns.barplot(x='Category',y='Discount',hue='Region',data=sort)
plt.title('Discount and Category')
```

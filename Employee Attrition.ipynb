# Data Visualization

```python
#Attrition:
attr=sns.countplot(x='Attrition',data=df)

for i in attr.containers:
    attr.bar_label(i)
```

```python
#Attrition with Gender:
attr=sns.countplot(x='Attrition',hue='Gender',data=df)

for i in attr.containers:
    attr.bar_label(i)
```

```python
#Business Travel:
count=sns.countplot(x='BusinessTravel',hue='Gender',data=df)

for i in count.containers:
    count.bar_label(i)
```

```python
# Condition Statement:
#1.To find the details of employee under attrition having 5+ experience in between the age of group 27-35(where age and TotalWorkingYears):

fil1=df.query('TotalWorkingYears>=5 & (Age > 27 & Age < 35)').groupby('Age',as_index=False)['TotalWorkingYears'].sum().sort_values(by='TotalWorkingYears',ascending=False)

fil1_count=sns.barplot(x='Age',y='TotalWorkingYears',data=fil1)

for i in fil1_count.containers:
    fil1_count.bar_label(i)
```

*The drawn from the graph is that employees with 5+ years of experience predominantly fall within the age range of 27 to 35, with 34-year-olds being the most prevalent.*

```python
#2.Employee having max salary working in differnt department who received less than 13% salary hike:
maxi=df.query('PercentSalaryHike < 13').groupby('Department',as_index=False)['MonthlyIncome'].max().sort_values(by='MonthlyIncome',ascending=False)

maxi_count=sns.barplot(x='Department',y='MonthlyIncome',data=maxi)

for i in maxi_count.containers:
    maxi_count.bar_label(i)
```

*The Research & Development department houses employees with the highest salaries, and notably, those receiving less than a 13% salary hike hold the top positions in terms of compensation.*

```python
#3.Employee having min salary working in differnt department who received less than 13% salary hike:
maxi=df.query('PercentSalaryHike < 13').groupby('Department',as_index=False)['MonthlyIncome'].min().sort_values(by='MonthlyIncome',ascending=False)

maxi_count=sns.barplot(x='Department',y='MonthlyIncome',data=maxi)

for i in maxi_count.containers:
    maxi_count.bar_label(i)
```

*The conclusion is that the Research & Development department comprises employees with the lowest salaries who received less than a 13% salary hike.*

```python
#4.Calculate the average salary of all gender who worked more than 3years whose education background is medical:
avgi=df.query("TotalWorkingYears>3 & EducationField=='Medical'").groupby('Gender',as_index=False)['MonthlyIncome'].mean().sort_values(by='MonthlyIncome',ascending=False)

avgi_count=sns.barplot(x='Gender',y='MonthlyIncome',data=avgi)
```

```python
#5.Total no.of.male and female under the PerformanceRating whose the marital status is married and have'nt received promotion in the last 2 years
pr=df.query("MaritalStatus=='Married' & YearsSinceLastPromotion==2").groupby('Gender',as_index=False)['PerformanceRating'].sum().sort_values(by='PerformanceRating',ascending=False)

pr_count=sns.barplot(x='Gender',y='PerformanceRating',data=pr)

for i in pr_count.containers:
    pr_count.bar_label(i)
```

*Males, with marital status as married and no promotion in the last 2 years, exhibit higher performance ratings compared to females based on the graph.*

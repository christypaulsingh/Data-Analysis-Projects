# Data Cleaning

```python
#Missing data/ Check the null value or empty values:
total=df.isnull().sum().sort_values(ascending=False)
percent=(df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)

missing_data=pd.concat([total,percent],axis=1,keys=['Total','Percent'])
missing_data
```

```python
#Drop the values:
df.drop(['Status', 'unnamed1', 'User_ID'] , axis=1 ,inplace=True)
```

```python
df
```

```python
#fillna is specific columns:
df['Amount'].fillna(df['Amount'].mean(),inplace=True)
```

```python
df.head(20)
```

```python
#Change data type in amount:
df['Amount']=df['Amount'].astype('int')
```

```python
#Duplicates:
df.duplicated().sum()
```

```python
#Remove the duplicated values:
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

*Complete the data cleaning process remove all duplicates values and null values.Finally total number of rows and columns is 11243 rows and 12 columns*

# Data Visulization

```python
#Gender:
#1.Count of gender:
gender_count=sns.countplot(x='Gender',data=df)

for i in gender_count.containers:
    gender_count.bar_label(i)   
```

*Count of gender is more females shopping the product when compared to males.*

```python
#2.Which gender is more order the product/items:
gender1=df.groupby(['Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

gender1_count=sns.barplot(x='Gender',y='Orders',data=gender1)

for i in gender1_count.containers:
    gender1_count.bar_label(i)
```

*Females is more orders the product when compared to males.*

```python
#3.Which gender is more spending the money:
sns.set(rc={'figure.figsize':(5,5)})
gender3=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

gender3_count=sns.barplot(x='Gender',y='Amount',data=gender3)
```

*Females is more amount the product when compared to males.*

```python
#Customers:
#Which customers is more orders the product/items on top 10 customers:
cust1=df.groupby('Cust_name',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
cust1_count=sns.barplot(x='Cust_name',y='Orders',data=cust1)

for i in cust1_count.containers:
    cust1_count.bar_label(i)
```

```python
#States:
#Which states is more orders the product/items on top 10:
states=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,10)})
states_count=sns.barplot(x='State',y='Orders',data=states)

for i in states_count.containers:
    states_count.bar_label(i)
```

```python
#Occupation:
#1.Which occupation is more spending the money for order in top 10 :
occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

occu_count=sns.barplot(x='Occupation',y='Amount',data=occupation)
```

```python
#Marital:
#Which marital/non-marital person is more orders the product/items:
marital=df.groupby(['Marital_Status'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(5,5)})
marital_count=sns.barplot(x='Marital_Status',y='Orders',data=marital)

for i in marital_count.containers:
    marital_count.bar_label(i)
```

```python
#Age group:
#1.count of age group:
sns.set(rc={'figure.figsize':(10,5)})
age=sns.countplot(x='Age Group',data=df)

for i in age.containers:
    age.bar_label(i)
```

```python
#2.count of age group in gender:
sns.set(rc={'figure.figsize':(10,5)})
age=sns.countplot(x='Age Group',hue='Gender',data=df)

for i in age.containers:
    age.bar_label(i)
```

```python
#3.Which age group is spending the more amount for shopping:
age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})

age_count=sns.barplot(x='Age Group',y='Amount',data=age)
```

```python
#Zone:
#Count of zone:
sns.set(rc={'figure.figsize':(10,5)})
zone=sns.countplot(x='Zone',hue='Gender',data=df)

for i in zone.containers:
    zone.bar_label(i)
```

```python
#which zone is more orders the product:
zone=df.groupby(['Zone'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(10,5)})
zone_count=sns.barplot(x='Zone',y='Orders',data=zone)

for i in zone_count.containers:
    zone_count.bar_label(i)
```
